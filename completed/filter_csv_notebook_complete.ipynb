# Filter a CSV

So we know how to [download a CSV from the Interent and read the data](read_csv_notebook_complete.ipynb) using the `csv` library.

Now we're ready to do something a bit more useful using these libraries, combined with the basics we learned on Day 1.

Once again, we'll start by importing the library code for downloading a file and processing a CSV:

```python
import csv
from urllib.request import urlretrieve
```

We're going to download a csv file. What should we name it?

```python
downloaded_file = "banklist.csv"
```

Now we can download the file using `urlretrieve`. Don't forget that it takes two arguments:

- the URL
- the name we'll give the file when we save it locally

```python
urlretrieve("https://s3.amazonaws.com/datanicar/banklist.csv", downloaded_file)
```

The output shows we successfully downloaded the file and saved it.

For this exercise, we'll:

- read the data
- filter the data for banks in California and save them in a list
- save the California banks list to a new CSV file

> NOTE: Below, we use the newline argument when we open the file to write so it doesn't add [additional, blank rows on Windows machines](https://stackoverflow.com/questions/3348460/csv-file-written-with-python-has-blank-lines-between-each-row/3348664).

```python
ca_banks = []
```

We'll use the `csv` library's [writer](https://docs.python.org/3/library/csv.html#csv.writer) functionality to...well...write data to a file by passing in the name of the new file as the first argument and delimiter as the second.

Then we'll use `reader` to open the file and see what is inside.

We specify the name of the file we just created, and we add a setting so we can open and read almost any CSV file.

```python
# open our downloaded file
with open(downloaded_file, 'r') as file:

    # Create a counter
    count = 0

    # Read the data
    reader = csv.reader(file)
    
    import pdb
    # loop through each row of the csv
    for row in reader:
        # Save the header row
        if count == 0:
            ca_banks.append(row)

        # Save the row if it's a CA bank
        if row[2].strip().upper() == 'CA':
            # Increment the count
            count += 1
            # Save the row to a our list
            ca_banks.append(row)
```

How do we find out how many banks failed in California?

```python
len(ca_banks)
```

And we can create a new CSV with just the CA banks.

```python
with open('ca_banks.csv', 'w') as output:
    writer = csv.writer(output)
    for row in ca_banks:
        writer.writerow(row)
```
